```python
import requests
from bs4 import BeautifulSoup
import csv
import time

# 검색어 설정
search_query = "k-pop"  # 여기에 검색어를 입력하세요.

# 총 페이지 수 설정
total_pages = 10
start_year = 2021
end_year = 2024

# CSV 파일 열기
csv_file = open('k-pop(2021-2023).csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['제목', '저자', '발행연도', '학회지'])  # CSV 파일의 헤더 작성

for page_year in range(start_year, end_year):
  for page in range(total_pages):
      start_count = page * 10

      search_url = f"https://www.riss.kr/search/Search.do?isDetailSearch=N&searchGubun=true&viewYn=OP&queryText=&strQuery={search_query}&exQuery=pyear%3A{page_year}%E2%97%88&exQueryText=%EB%B0%9C%ED%96%89%EC%97%B0%EB%8F%84+%5B2023%5D%40%40pyear%3A2023%E2%97%88&order=%2FDESC&onHanja=false&strSort=RANK&p_year1=&p_year2=&iStartCount={start_count}&orderBy=&mat_type=&mat_subtype=&fulltext_kind=&t_gubun=&learning_type=&ccl_code=&inside_outside=&fric_yn=&db_type=&image_yn=&gubun=&kdc=&ttsUseYn=&l_sub_code=&fsearchMethod=search&sflag=1&isFDetailSearch=N&pageNumber=1&resultKeyword=k-pop&fsearchSort=&fsearchOrder=&limiterList=&limiterListText=&facetList=&facetListText=&fsearchDB=&icate=re_a_kor&colName=re_a_kor&pageScale=10&isTab=Y&regnm=&dorg_storage=&language=&language_code=&clickKeyword=&relationKeyword=&query={search_query}"

      response = requests.get(search_url)
      time.sleep(2)

      soup = BeautifulSoup(response.text, 'html.parser')

      # 논문 정보 추출
      papers = soup.select('.srchResultListW .cont')
      for paper in papers:
          title = paper.select_one('.title').get_text(strip=True)
          author = paper.select_one('.writer').get_text(strip=True)
          year = page_year
          journal = paper.select_one('.assigned').get_text(strip=True)

          # CSV 파일에 작성
          csv_writer.writerow([title, author, year, journal])

# CSV 파일 닫기
csv_file.close()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
from matplotlib import font_manager, rc
import pandas as pd

# Colab에서 한글 폰트 설정
#!apt-get update -qq
#!apt-get install fonts-nanum* -qq
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# CSV 파일 읽어오기
df = pd.read_csv('k-pop(2021-2023).csv', encoding='utf-8')

# 제목 열에서 텍스트 가져오기
text = ' '.join(df['제목'])

# WordCloud 생성
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color = 'white',
    font_path = font_path,
).generate(text)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
